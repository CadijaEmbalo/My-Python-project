```python
#If a number is divisible by 3, print "fizz."
if i%3==0:
    print("fizz")
```

```python
#If a number is divisible by 5, print "buzz."
if i%5==0:
    print("buzz")
```

```python
#If a number is both divisible by 3 and by 5, print "fizzbuzz."
if i%3==0 and i%5==0:
    print("fizzbuzz")
```

```python
for i in range(1,101):
  if i%3==0 and i%5==0:
    print("fizzbuzz")
  elif i%3==0:
    print("fizz")
  elif i%5==0:
    print("buzz")
  else:
    print(i)
```
